```python
#CSVファイルを辞書形式で読み込む
# https://esu-ko.hatenablog.com/entry/2016/02/16/Python%E3%81%A7CSV%E3%83%95%E3%82%A1%E3%82%A4%E3%83%AB%E3%82%92%E6%89%B1%E3%81%86%28%E3%83%AA%E3%82%B9%E3%83%88%E3%82%84%E8%BE%9E%E6%9B%B8%E3%81%A8%E3%81%AE%E3%82%A4%E3%83%B3%E3%82%BF%E3%83%BC%E3%83%95
import csv
import pprint

songs = []

with open('songs.csv', 'r',encoding="utf-8") as f:
    # DictReaderと共にHeaderを渡すことで辞書形式で返す。
    # 取得したデータを1行ずつ出力。
    reader = csv.DictReader(f)
    for row in reader:
        songs.append(row)
pprint.pprint(songs)
```

```python
#名寄せのために、所属を削除
import re
songs_text = json.dumps(songs, indent=4, ensure_ascii=False)
songs_text = re.sub('[（\(]Arte Refact[）\)]', '',songs_text)
songs_text = re.sub('[（\(]Nostalgic Orchestra[）\)]', '',songs_text)
songs_text = re.sub('[（\(]Void_Chords[）\)]', '',songs_text)
songs_text = re.sub('[（\(]Dream Monster[）\)]', '',songs_text)
songs_text = re.sub('[（\(]fhana[）\)]', '',songs_text)
songs_text = re.sub('NBSI[\(（]([^\)）]*)[\)）]', '\\1',songs_text)
songs_text = re.sub('BNSI[\(（]([^\)）]*)[\)）]', '\\1',songs_text)
songs = json.loads(songs_text)
```

```python
#スコア計算
import collections

for song in songs:
    artists = []
    artists.append(song['作詞'])
    artists.append(song['作曲'])
    artists.append(song['編曲'])
    artists.append(song['Idol1'])
    artists.append(song['Idol2'])
    artists.append(song['Idol3'])
    artists.append(song['Idol4'])
    artists.append(song['Idol5'])
    #アーティスト名にカンマが入っていたら分離
    for i in range(len(artists)):
        artist = artists[i]
        sep = artist.find('、')
        if sep>0:
            splitted = artist.split('、')
            artists[i] = ''
            artists.extend(splitted)
    
    #個数を数える
    c = collections.Counter(artists)
    c.pop('', None)
    song['score'] = dict(c)
    scores = []
    for key in c.keys():
        scores.append({'artist': key, 'value':c[key]})
    song['scores'] = scores
    pprint.pprint(song)
```
